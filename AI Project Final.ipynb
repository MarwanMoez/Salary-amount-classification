```python
categorical_df=['work-class','marital-status','position','relationship','race','native-country']
categorical_test=['work-class','marital-status','position','relationship','race','native-country']
```

```python
def onehot(df,category):
    for col in category:
        dummies=pd.get_dummies(df[col])
        df=pd.concat([df,dummies],axis=1)
        df.drop(col,axis=1,inplace=True)
    return df
```

```python
df=onehot(df,categorical_df)
```

```python
df
```

```python
df['sex'].replace(' Male',1,inplace=True)
df['sex'].replace(' Female',0,inplace=True)
df['salary'].replace(' >50K',0,inplace=True)
df['salary'].replace(' <=50K',1,inplace=True)
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train=df.drop(['salary'],axis=1)
y_train=df['salary']
```

```python
n_estimators_range=[int(x) for x in np.linspace(10,200,15)]
max_depth=[int(x) for x in np.linspace(10,200,15)]
```

```python
param_grid={'n_estimators':n_estimators_range,
           'max_depth':max_depth}
print(param_grid)
```

```python
rfc=RandomForestClassifier()
```

```python
rfc_random=RandomizedSearchCV(estimator=rfc,param_distributions=param_grid,cv=5)
```

```python
rfc_random.fit(X_train,y_train)
```
